```python
%matplotlib inline

from collections import defaultdict
from csv import reader
from os import listdir
from os.path import join
```

```python
DATA_FOLDER = "/Users/adrienfallou/Documents/code_perso/discord_activity/out"
```

```python
daily_activity = dict()
for filename in listdir(join(DATA_FOLDER, "day_activity")):
    with open(join(DATA_FOLDER, "day_activity", filename)) as f:
        day_timestamp = int(filename.split(".")[0])
        csv_reader = reader(f)
        day_activity = {row[0]: int(row[1]) for row in csv_reader}
        daily_activity[day_timestamp] = day_activity
```

```python
days = list(daily_activity.keys())
user_ids = set()
for _, day_activity in daily_activity.items():
    day_users = set(day_activity.keys())
    user_ids = user_ids.union(day_users)
```

```python
user_activity = {
    user_id: [0] * (len(days))
    for user_id
    in user_ids
}
for day_index, day in enumerate(days):
    for user_id, message_count in daily_activity[day].items():
        user_activity[user_id][day_index] = message_count
```

```python
total_msgs = {
    user_id: sum(msg_counts)
    for user_id, msg_counts
    in user_activity.items()
}
total_msgs
```
